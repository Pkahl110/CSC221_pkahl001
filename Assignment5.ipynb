```python
# Factorial

def factorial(n):

    product = 1

    numbers = range(1, n+1)

    for x in numbers:
        product *= x

    return product

print(factorial(5))
print(factorial(10))
print(factorial(20))
print(factorial(50))
```

```python
#3

def special_factorial(n):

    product = 1

    numbers = list(range(1, n+1))

    if 4 in numbers:
        numbers.remove(4)
    if 13 in numbers:
        numbers.remove(13)
    if 22 in numbers:
        numbers.remove(22)

    for x in numbers:
        product *= x

    return product

n = 25
print(special_factorial(5))
print(special_factorial(10))
print(special_factorial(20))
print(special_factorial(50))
```
